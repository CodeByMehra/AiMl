```python
import json  
```

```python
#load the data
def load_data(filename):
    with open (filename, "r") as f:
        data = json.load(f)

    return data
```

```python
data= load_data("01_store_data.json")
print(data)
```

```python
# Clean & structure the data
def clean_data(data):
    text_to_num = {"one": 1, "two": 2, "three": 3, "four": 4, "five": 5}
    cleaned_data = []
    unique_users = set()

    for user in data:
        # Clean ratings
        raw_rating = str(user["rating"]).strip().lower()
        if raw_rating in text_to_num:
            raw_rating = text_to_num[raw_rating]
        user["rating"] = raw_rating

        # Handle missing vals
        raw_age = user.get("age")
        if raw_age == None:
            user["age"] = None

        # Deduplication
        if user["name"].strip() in unique_users:
            continue

        unique_users.add(user["name"])
        cleaned_data.append(user)

    return cleaned_data
```

```python
clean_data(data)
```

```python
#Get meaningfuu insights from data
def get_insights(data):
    total_rating = 0 
    for user in data:
        total_rating = total_rating + float(user["rating"])

    print(f"average rating = {total_rating/len(data)}")

    #percentage of uusers with poor rating
    poor_ratings = 0
    for usuer in data:
        if (float(user["rating"]) < 3):
            poor_ratings = poor_ratings +  1

    print(f"Poor ratings: {poor_ratings/len(data) *100 }%")
```

```python
get_insights(data)
```

```python
#recommendations based on rating

def get_recommendations(data):
    recommendations = []

    for user in data:
        curr_recommendation = {}
        curr_recommendation["name"] = user ["name"]
        
        if (float(user["rating"]) >= 4):
            curr_recommendation["brand"] = "Apple"
        else:
            curr_recommendation["brand"] = "Samsung"

        recommendations.append(curr_recommendation)
    return recommendations
```

```python
get_recommendations(data)
```
