```python
import pandas as pd
df = pd.read_csv("raw_data.csv")
df
```

```python
#handle missing values:
df.isnull()             #we can also use isna()
df.isnull().sum()        #count nmber of missing vales per column
df.dropna()              #to drop rows having any null value
df.fillna(0)            #to fill null values as given value
```
