```python
#Filtering of data
import pandas as pd
df = pd.read_csv("globalAirQuality.csv")
df
```

```python
#cities where aqi is greater than 100:

df[df["aqi"] > 100]
```

```python
#cities where aqi is greater than 100 and temp ht than 30:

df[(df["aqi"] > 100) & (df["temperature"] > 30)] 
```

```python
#Query method: 
df.query("aqi > 150")
```
